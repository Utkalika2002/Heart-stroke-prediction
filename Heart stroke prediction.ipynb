## Label encoding

```python
from sklearn.preprocessing import LabelEncoder
obj_cols=['gender', 'ever_married', 'Residence_type','stroke']

lb=LabelEncoder()

for i in obj_cols:
    print(f"Value Counts of column:{i}")
    print(data.loc[:,i].value_counts())
    data[i] = lb.fit_transform(data[i])
    print("The classes which labeled from 0 are: ",lb.classes_,end="\n===========================================\n")
```

```python
data
```

```python
data=pd.get_dummies(data)
```

```python
data
```

```python
data.columns
```

## Splitting data into dependent & independent variables

```python
x_data = data.drop("stroke" , axis = 1).values
y_data = data["stroke"]
```

## Traing & Testing the data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train , y_test = train_test_split(x_data,y_data, test_size=0.25 )
```

## K Nearest Neighbour(KNN) Classifier

```python
from sklearn.neighbors import KNeighborsClassifier  
classifier= KNeighborsClassifier(n_neighbors=6)  
classifier.fit(x_train, y_train)  
print("Accuracy on Traing set: ",classifier.score(x_train,y_train))
print("Accuracy on Testing set: ",classifier.score(x_test,y_test))
```

## Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(random_state=10,max_depth=3)
classifier.fit(x_train,y_train)
print("Accuracy on Traing set: ",classifier.score(x_train,y_train))
print("Accuracy on Testing set: ",classifier.score(x_test,y_test))
```

## Naive Bayes Classifier 

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
print("Accuracy on Traing set: ",gnb.score(x_train,y_train))
print("Accuracy on Testing set: ",gnb.score(x_test,y_test))
```

### Our KNN classifier model gives accuracy of 94.44% which is better among all models.Hence we select KNN classifier.
